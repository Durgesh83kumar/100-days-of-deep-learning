```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(oov_token = "<oov>") # Use an OOV token in the tokenizer so that unseen words are mapped to a fixed index instead of breaking the model.
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
print(tokenizer.word_index)
```

```python
len(tokenizer.word_index)
```

```python
vocab_size = len(tokenizer.word_index) + 1
print(vocab_size)
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1,len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding = 'pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
print(X.shape)
print(y.shape)
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=vocab_size) # here one hot encoding is getting done
```

```python
y
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()

model.add(Embedding(input_dim = vocab_size,output_dim=100,input_shape=(16,)))
model.add(LSTM(150))
model.add(Dense(vocab_size,activation='softmax'))

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'],label='accuracy')
plt.plot(history.history['val_accuracy'],label='val_accuracy')
```

```python
import time
import numpy as np
text = "how"

for i in range(10):
    # tokenize
    token_text = tokenizer.texts_to_sequences([text])[0]
    # padding
    padded_token_text = pad_sequences([token_text],maxlen=16,padding='pre')
    # predict
    pos = np.argmax(model.predict(padded_token_text)) # argmax() will tell that at which position maximum probability occur

    for word,index in tokenizer.word_index.items():
        if index == pos:
            text = text + " " + word
            print(text)
            time.sleep(0.5)
```
