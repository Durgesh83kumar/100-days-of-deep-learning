```python
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('D:/deep_learning/11_Data_augmentation/Data_aug/cat.jpg',target_size=(200,200))
```

```python
img
```

```python
type(img) # image is not in numpy array
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip =True,
    width_shift_range = 0.2,
    height_shift_range=0.2,
    fill_mode = 'constant'
)
```

```python
img = image.img_to_array(img)
```

```python
type(img) # image is converted in numpy array
```

```python
img.shape
```

```python

input_batch = img.reshape(1,200,200,3)
```

```python
i = 0

for output in datagen.flow(input_batch,batch_size=1,save_to_dir='Data_aug'): # flow is used when working on single image
    # flow_from_directory is used for all the images contained in a folder

    i = i + 1

    if i==10:
        break
```
