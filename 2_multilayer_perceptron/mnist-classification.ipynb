```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
X_train[0] # the value  of each image is from 0 to 255 therefor isko 0 se 1 ke beech m lane k liye divide by 255 karna padega
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
X_train[0]
```

### Flatten() :
Flatten reshapes input, does not learn parameters

Converts multi-dimensional input → 1D

Commonly used after image input

If your input is an image of size 28 × 28:

So the output becomes a 1D array of length 784.

### Why we need Flatten()

Dense (Fully Connected) layers cannot accept 2D input.
They only accept 1D vectors.

So before a Dense layer, we use Flatten.

```python
model = Sequential()

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax')) # output layer m utne hi node lete h jitne classes hote h
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy']) # here we use 'sparse_categorical_crossentropy' instead categorical_crossentropy because in sparse_categorical_crossentropy we don't need to one-hot-encoding with our label
```

```python
history = model.fit(X_train,y_train,epochs=20,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
